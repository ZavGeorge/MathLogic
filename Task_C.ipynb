```python
def brackets_in_line(line):
    line = parser(line)
    while ',' in line:
        i = line.find(',')
        if line[i-1] == '|' or line[i-1] == '&':
            symbol = line[i-1]
        else:
            symbol = '->'
        j = i+1
        c_l = 0
        c_r = 0
        while True:            
            if line[j] == '(':
                c_l +=1
            if line[j] ==')':
                c_r +=1
            if line[j] == ',' and c_l == c_r:
                break
            j +=1
        left_elem = line[i+1:j]
        print(left_elem)
#         if left_elem[1] == '!':
#             left_elem = left_elem[1:len(left_elem) -1]
#         print(left_elem)
        k = j
        c_l = 0
        c_r = 0
        while k < len(line):      
            if line[k] == '(':
                c_l +=1
            if line[k] ==')':
                c_r +=1
            if line[k] == ')' and c_l == c_r:
                break
            k +=1
        right_elem = line[j+1:k] 
#         print(right_elem)
#         if right_elem[1] == '!':
#             right_elem = right_elem[1:len(right_elem) -1]
        
        line = line.replace(symbol + ',' + left_elem + ',' + right_elem, left_elem + symbol + right_elem ,1)

    return line
        
            
    
```
